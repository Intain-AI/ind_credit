```python
from pymongo import MongoClient
IND_CREDIT_MONGODB_PORT = 27017
IND_CREDIT_MONGODB_PORT = 27017
IND_CREDIT_MONGODB_URL = 'localhost'
IND_CREDIT_MONGODB_NAME = 'ind_credit_db'
```

```python
uri = "mongodb://{}:{}/{}".format(IND_CREDIT_MONGODB_URL, IND_CREDIT_MONGODB_PORT,IND_CREDIT_MONGODB_NAME)
# uri = "mongodb+srv://12345:12345@cluster0.thhvq.mongodb.net/myFirstDatabase?retryWrites=true&w=majority"
mongo_db_client = MongoClient(uri)
db = mongo_db_client[IND_CREDIT_MONGODB_NAME]

collection_job = db['job_detail']
```

```python

response=collection_job.find({},{
            '_id': 0,
            "job_size" : 0,
            'upload_date_time':0,
            'emailid':0,
            'channel':0,
            'reviewed_by':0,
            'document_name':0,
            'folder_path' :0,
            'excel_file_path':0,
            'final_excel_path':0
        })
```

```python
import pandas as pd
df=pd.DataFrame()
```

```python
for i in response:
    # print(i)
    df=df.append(i,ignore_index=True)
```
